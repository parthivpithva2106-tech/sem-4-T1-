```python
url=f"http://api.openweathermap.org/data/2.5/forecast?lat={latitude}&lon={latitude}&appid={api_key}"
res=req.get(url)
print(res)
```

```python
fiveday=res.json()
pprint.pprint(fiveday)
d={"DataTime":[],"Temp":[],"pressure":[],"Humidity":[],"weather":[]}
for i in fiveday['list']:
    d["DataTime"].append(i["dt_txt"])
    d['Temp'].append(i['main']['temp'])
    d['pressure'].append(i['main']['pressure'])
    d['Humidity'].append(i['main']['humidity'])
    d['weather'].append(i['weather'][0]['description'])
import pandas as pd
df=pd.DataFrame(d)
print(df)
        
    
```
