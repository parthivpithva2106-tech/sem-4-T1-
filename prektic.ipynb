3)Apply PolynomialFeatures (degree = 2)
4)Train model using LinearRegression
5)Predict score for Hours = 12
6)Plot graph (actual vs predicted curve)

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
#1)Load the dataset using pandas
df=pd.read_csv('student_growth.csv')
df
#2)Separate input (X) and output (Y) 
x=df[['hours']]
y=df['score']
print(x.shape)
print(y.shape)
#3)Apply PolynomialFeatures (degree = 2) 
poly=PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)
#4)Train model using LinearRegression 
model = LinearRegression()
model.fit(x_poly,y)
print(model.coef_)
print(model.intercept_)
#5)Predict score for Hours=12
leval=np.array([[12]])
leval_poly=poly.fit_transform(leval)
modelpredit=model.predict(leval_poly)
print(modelpredit)

```
