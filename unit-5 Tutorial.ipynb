## classfication model
- **Data:** Divided into predefined, discrete categories
- **Goal:** Assign as input to one of the categories
-  **Model output:** A specific categories from a set 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier,plot_tree

df=pd.read_csv('https://raw.githubusercontent.com/patelmanishv/Sem4Data/refs/heads/main/Datasets/job_offers.csv')
df

#Define features and target
x=df.drop(columns=['JobOffered'])
y=df['JobOffered']
dtree=DecisionTreeClassifier(criterion='entropy',max_depth=3)
dtree.fit(x,y)


predictions=dtree.predict(x)


#plot the decision tree
plt.figure(figsize=(12,8))
plot_tree(dtree,filled=True,feature_names=['CGPA',	'Communication',	'Aptitude',	'ProgrammingSkill'],class_names=['No','Yes'])
plt.title("decision tree")
plt.show()

#only 
#dtree??



```

## EVALUTION METRICS
- CONFUSION MATRIX
- ACCURACY SCORE
- PRECISION SCORE
- RECALL SCORE
- FA SCORE
- CLASSIFICATION REPORT

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,classification_report
```

```python
cm=confusion_matrix(y,predictions)
print('confusion_matrix=\n',cm)
```

```python
TN,FP,FN,TP=cm.ravel()
print(TN)
print(FP)
print(FN)
print(TP)
```

```python
#confusion_matrix??
```

## Accuracy score


```python
accuracy=(TP+TN)/(TP+TN+FP+FN)
print(accuracy)
```

## precision score
- precision =TP/(TP+FP)

```python
precision =TP/(TP+FP)
print(precision)
```

## RECALL

```python
recall=TP/(TP+FN)
print(recall)
```

## f1 score

```python
f1=2*(precision*recall)/(precision+recall)
print(f1)
```

```python
print(classification_report(y,predictions))
```
