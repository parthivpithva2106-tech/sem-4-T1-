## tokens

```python
import  requests as req
url="https://jsonplaceholder.typicode.com/posts"
token="mysecrettoken"
headers={
    "Authorization":f"Bearer{token}"}
res=req.get(url,headers=headers)
print(res.status_code)
print(res.text)
```

## PAGINATION
- IT IS USED TO PAGE CONTROL THE PAGE FLOW OF THE RESOURCE REQ
- for ofline 3 mark 

```python
url="https://jsonplaceholder.typicode.com/posts"
page=1
while page <=3:
    params={
        "_page":page,
        "_limit":5
    }
    res=req.get(url,params=params)
    print(res.status_code)
    data=res.json()
    for post in data:
        print(post['id'],post['title'])
        page+=1
```

## rate Limit
- api providers use limit to control the api usage.the limit will restrict further usage of api 
until reset

### check rate limit of github api


```python
import requests as req
res=req.get("https://api.github.com/rate_limit")
data=res.json()
print("limit-->",data['rate']['limit'])
print("used-->",data['rate']['used'])
print("left-->",data['rate']['remaining'])
print("reset-->",data['rate']['reset'])
```

### check the rate limit by fetching the page


```python
for i in range(60):
    res=req.get('https://api.github.com/octocat')
    print(i+1,res.status_code)
    if res.status_code==403:
        data=res.json()
        if "API rate limit exceeded" in data.get("message",""):
            print("rate limit exceeded")
            print(data['message'])
            break
```

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
with open("Certified used Mercedes-Benz for sale.html",encoding='utf-8') as file:
    html=file.read()
soup=BeautifulSoup(html,"html.parser")
print(soup)
    
```

```python
item=soup.find_all('div',class_="vehicle-card")
print(item)
```

```python
deta=[]
for items in item:
    name=items.find("h2",class_="title").get_text(strip=True)
    mile=items.find("div",class_="mileage").get_text(strip=True)
    price=items.find("span",class_="primary-price").get_text(strip=True)
    deler_name=items.find("div",class_="dealer-name").get_text(strip=True)
    review=items.find("span",class_="sds-rating__link sds-button-link").get_text(strip=True).split()[0][1::]
    
    deta.append({
        'Name':name,'Mileage':mile,'Price':price,'Deler_name':deler_name,'reviews':review
    })
df=pd.DataFrame(deta)
```
